
• Solicite el precio de 5 producto y para cada producto generarle
el IVA del 19%

• Generar el valor base de todos los productos

• Generar el valor total del IVA

• Generar el valor total de la Factura (valor base + IVA).

```python
suma=0
for i in range(0, int(input("Ingrese la cantidad de productos a comprar: "))):
  suma+=int(input("Ingrese el valor del producto num "+str(i)+": "))
print("El precio total de todos los productoes es : "+str(suma))
IVA=suma*0.19
Total=suma+IVA
print("El iva total de los productos es de:", IVA)
print("El total a pagar con el iva incluido es de:", Total)
```
